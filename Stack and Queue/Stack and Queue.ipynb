```python
# Implementation of stacks using classes - arrays/lists

class Stack:
    
    def __init__(self):
            self.items = [] # this is an empty list

    def isEmpty(self):
        return self.items == []

    def push(self, item):
        self.items.append(item)
        print("elements pushed to the stack",item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[len(self.items)-1]

    def size(self):
        return len(self.items)
    
    def PrintStack(self):
        print(self)
        

    
s = Stack()
s.push(1)
s.push(2)
# this only prints the location of stack
# print(s)
s.push('p')
s.push('e')
# print(s)
# peek prints the value of the TOPmost element in the Stack
print(s.peek())
# returns False if the Stack is not empty
print(s.isEmpty())
# s.PrintStack()
# print(s.get())

# Implementation using TOP Pointer
# https://www.geeksforgeeks.org/print-stack-elements-from-top-to-bottom/
```

```python
# Linked List Representation of Stacks

class Node:
    def __init__(self,data = None):
        self.data = data
        self.next = None
        
class SinglyRepOfStacks:
    def __init__(self):
        self.head = None
        # head is a pointer to the first node, or it holds the address of the first node
        
    def push(self,data):
        # prepend
        current_node = self.head
        new_node = Node(data)
        self.head = new_node
        new_node.next = current_node
        
    def pop(self):
        # predelete
        current_node = self.head
        self.head = current_node.next
        
    def peek(self):
        current_node = self.head
        print(current_node.data)

    def print_list(self):
        current_node = self.head
        while(current_node):
            print(current_node.data)
            current_node = current_node.next


s = SinglyRepOfStacks()
s.push(1)
s.push(2)
s.push(3)
s.push(4)
s.push(5)
s.push(6)
s.pop()
s.peek()
s.print_list()
```

```python
# Linked List representation of Queues

class Node:
    def __init__(self,data=None):
        self.data = data
        self.next = None
        
class LinkedListRepOfQueue:
    def __init__(self):
        self.head = None
        
    def insert(self,data):
        # append
        new_node = Node(data)
        
        if not self.head:
            self.head = new_node
            return
        
        current_node = self.head
        
        while(current_node):
            if(current_node.next == None):
                current_node.next = new_node
                new_node.next = None
            current_node = current_node.next
        
    def delete(self,data):
        # predelete
        current_node = self.head
        self.head = current_node.next
        
    def print_list(self):
        current_node = self.head
        while(current_node):
            print(current_node.data)
            current_node = current_node.next
        
        
q = LinkedListRepOfQueue()
q.insert(1)
q.insert(5)
q.insert(2)
q.insert(7)
q.delete(1)
q.print_list()
```

```python
# Implementation of Circular Queue

class CircularQueue:

    #Constructor
    def __init__(self):
        # queue is defined as a list here
        self.queue = list()
        self.head = 0 # Head is itself the FRONT pointer here
        self.tail = 0 # Tail is itself the REAR pointer here
        # predefined legth of the circular queue ?
        self.maxSize = 8

    #Adding elements to the queue
    def enqueue(self,data):
        if self.size() == self.maxSize-1:
            return ("Queue Full!")
        # if the size of the list is full or if maxsize-1, length of the list - 1,  then queue is full
        self.queue.append(data)
        self.tail = (self.tail + 1) % self.maxSize # incremet the REAR pointer to the new element position 
        return data

    #Removing elements from the queue, predelete
    def dequeue(self):
        if self.size()==0:
            return ("Queue Empty!") # Both FRONT and REAR will be pointing to 0th cell only
        data = self.queue[self.head] # data will hold the first element that is to be deleted ..
        self.head = (self.head + 1) % self.maxSize # self.head or head pointer will point to the second ele at 1st location,
        # dropping the first element at 0th index ,dropping the element in 0th position (0+1)% 3 = 1 (remainder)
        # here mod is used to round the size of the circular queue after the max size of the queue in order to fill the
        # queue if there are empty spaces at the beginning of the queue
        return data # element at 0th location is dropped and is returned

    #Calculating the size of the queue
    def size(self):
        if self.tail>=self.head:
            return (self.tail-self.head)
        return (self.maxSize - (self.head-self.tail))

q = CircularQueue()
print(q.enqueue(1))
print(q.enqueue(2))
print(q.enqueue(3))
print(q.dequeue())
print(q.dequeue())
print(q.enqueue(4))
print(q.enqueue(5))
print(q.dequeue())
print(q.dequeue())
print(q.dequeue())
print(q.dequeue())
```
