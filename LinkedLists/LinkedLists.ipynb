```python
# Implementation of Circular Singly Linked List

class Node:
    def __init__(self,data = None):
        self.data = data
        self.next = None
    
class CircularSinglyLinkedList:
    def __init__(self):
        self.head = Node(None) # another way to write it as a null pointer or self.head = None
        self.tail = Node(None) # shows that this tail does not contain data, but only a pointer
        self.head.next = self.tail # head will point to tail
        self.tail.next = self.head # tail will point to head


    def append(self,data):
        new_node = Node(data)
        
        if self.head.data is None:
            # if the list is empty, then both head and tail will point to the same node, and tail will again point to head
            self.head = new_node
            self.tail = new_node
            new_node.next = self.head
            print("The new node added at the end of the list is",new_node.data)
            
        else:
            self.tail.next = new_node
            self.tail = new_node
            self.tail.next = self.head
            print("The new node added at the end of the list is",new_node.data)
            
    
    def prepend(self,data):
        new_node = Node(data)
        
        if self.head.data is None:
            # if the list is empty, then both head and tail will point to the same node, and tail will again point to head
            self.head = new_node
            self.tail = new_node
            new_node.next = self.head
            
        else:
            # you have to use a temp variable here to store the original head
            temp = self.head
            self.head = new_node
            new_node.next = temp
            self.tail.next = self.head
            print("The new node added at the beginning is",new_node.data)
            
            
    def predelete(self,data):
        current_node = self.head
        if current_node.data == data:
            self.head = current_node.next
            self.tail.next = self.head
            print("The first node that is deleted from the list is", current_node.data)
            
            
    def postdelete(self):
        
        current_node = self.head
        prev_ptr = Node(None)
        while(current_node != self.tail):
            prev_ptr=current_node
            current_node=current_node.next
        self.tail = prev_ptr
        self.tail.next = self.head
        print("The last node that is deleted from the list is", current_node.data)

        
    def searching(self,data):
        current_node = self.head
        while(current_node):
            if current_node.data == data:
                print(current_node.next)
            current_node = current_node.next
            if current_node == self.tail:
                
                print("The node is not found in the list")
                break
        
        
    def count(self):
        temp = self.head
        count = 0
        if (temp != None):
            count += 1
            temp = temp.next
        while (temp != self.head):
            count += 1
            temp = temp.next
        return count 
            
        
        
    def printlist(self):
        current_node = self.head
        while(current_node):
            print(current_node.data)
            if current_node == self.tail:
                break
            current_node = current_node.next
            
        
        
my_list = CircularSinglyLinkedList()
my_list.append(5)
my_list.append(6)
my_list.append(7)
my_list.append(4)
my_list.prepend(2)
#my_list.prepend(4)
#my_list.prepend(1)
print("No. of nodes in the list before the implementation: ", my_list.count())
my_list.predelete(2)
my_list.postdelete()
my_list.searching(8)
# my_list.printlist()
print("No. of nodes in the list after the implementation: ", my_list.count())


        
```
